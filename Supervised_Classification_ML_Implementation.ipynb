```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
data['Gender']=label_encoder.fit_transform(data['Gender'])
```

We will implement the following algorithms on top of this data and check which algorithm gives us the least value of FP and FN

- Logistic Regression
- Decision Trees
- Random Forest

```python
x=data.drop(['Purchased'],axis=1) #independent columns
y=data['Purchased'] #dependent column
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
model_1=LogisticRegression()
```

```python
model_1.fit(x_train,y_train)
```

```python
predictions_1=model_1.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm_1=confusion_matrix(y_test,predictions_1)
ac_1=accuracy_score(y_test,predictions_1)
```

```python
#decision trees
from sklearn.tree import DecisionTreeClassifier
model_2=DecisionTreeClassifier()
```

```python
model_2.fit(x_train,y_train)
```

```python
predictions_2=model_2.predict(x_test)
```

```python
cm_2=confusion_matrix(y_test,predictions_2)
ac_2=accuracy_score(y_test,predictions_2)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
model_3=RandomForestClassifier()
```

```python
model_3.fit(x_train,y_train)
```

```python
predictions_3=model_3.predict(x_test)
```

```python
cm_3=confusion_matrix(y_test,predictions_3)
ac_3=accuracy_score(y_test,predictions_3)
```

```python
print(cm_1) #FP=6, FN=15
print(cm_2) #FP=6, FN=9
print(cm_3) #FP=7, FN=6
```

```python
print(ac_1,"accuracy obtained by Logistic Regression")
print(ac_2,"accuracy obtained by Decision Tree")
print(ac_3,"accuracy obtained by Random Forest")
```
